The short EDA of the set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re
import sys
import nltk
```

```python
train = pd.read_csv('../data/raw/train.csv')
test = pd.read_csv('../data/raw/test.csv')
train.head()
```

First, lets check a distribution of positive and negative reviews

```python
pd.Series(train['sentiment'].value_counts()).plot(kind = 'bar')
```

We can see that set is perfectly balanced so we dont need to balance it

Now lets check avg. length of the review

```python
train["review_length"] = train["review"].apply(len)
train["review_length"].plot(kind = 'hist', bins = 100)
```

We can see that reviews are rather long, avegeraly around 1000 letters.

Lastly, is it worth to see the most frequent word. For that I will use train_lemmatized.csv file which is generated by data_loader.py

```python
train_lemmatized = pd.read_csv('../data/processed/train_lemmatized.csv')
positive = train_lemmatized[train_lemmatized['sentiment'] == 1]
negative = train_lemmatized[train_lemmatized['sentiment'] == 0]
```

```python
pd.Series(' '.join(positive['review']).split()).value_counts()[:20].plot(kind = 'bar')
```

```python
pd.Series(' '.join(negative['review']).split()).value_counts()[:20].plot(kind = 'bar')
```
